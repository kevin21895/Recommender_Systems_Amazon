```python
def review_to_words(raw_review,nostring):
    review_text = BeautifulSoup(raw_review,nostring).get_text()
    letters_only = re.sub("[^a-zA-Z]"," ",review_text)
    words = letters_only.lower().split()
    stops = set(stopwords.words("english"))
    necessary_words = [w for w in words if w not in stops]
    return (" ".join(necessary_words))
```
